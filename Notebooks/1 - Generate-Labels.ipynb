## Generate labels

```python
df_genres.rename(columns={"#tracks":"track_id"},inplace=True)
```

```python
df_tracks_genres = pd.concat([df_raw_tracks, df_genres], axis=1, join="inner")
```

```python
df_tracks_genres.genre_id.value_counts()
```

```python
def create_labels(arr):
    labels = sorted(list(set(arr)))
    return dict([(label, i) for i, label in enumerate(labels)])   
```

```python
labels = {
        'global': create_labels(df_tracks_genres.genre_id)
}
```

```python
labels_path = os.path.join(metadata_path,"labels.json")
```

```python
with open(labels_path, "w+") as f:
    f.write(json.dumps(labels))
```
